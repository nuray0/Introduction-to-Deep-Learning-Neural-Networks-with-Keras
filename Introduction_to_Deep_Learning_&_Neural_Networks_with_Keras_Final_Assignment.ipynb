```python
import pandas as pd
import numpy as np
import keras

from keras.models import Sequential
from keras.layers import Dense

concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head(10)
```

```python
x = concrete_data.drop(columns = ['Strength'])
y = concrete_data['Strength']

n_cols = x.shape[1]
```

#A. Build a baseline model (5 marks) 

Use the Keras library to build a neural network with the following:

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.



```python
def reg_mod():
  model = Sequential()
  model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
  model.add(Dense(1))
  model.compile(optimizer = 'adam', loss = 'mean_squared_error')
  return model

model = reg_mod()
```

1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
print(x_train.shape, x_test.shape)
```

2. Train the model on the training data using 50 epochs.


```python
model.fit(x_train, y_train, epochs = 50)
```

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

```python
from sklearn.metrics import mean_squared_error
yhat = model.predict(x_test)
mean_squared_error(y_test, yhat)
```

```python
def evaluate_model(model, x, y, epochs):
  mse = []
  for i in range(50):
    # 1. split the data 
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)  
    # 2. train 50 epochs
    model.fit(x_train, y_train, epochs=epochs, verbose=0)
    # 3. mse
    yhat = model.predict(x_test)
    mse.append( mean_squared_error(y_test, yhat) )
    print('.', end='')
  print(' Done!')
  # return the mean and stddev of the mse list
  return np.mean(mse), np.std(mse)
```


4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

5. Report the mean and the standard deviation of the mean squared errors.

```python
model1 = reg_mod()
mean_mse, std_mse = evaluate_model(model1, x, y, 50)
# Report the mean and stddev of the mean squared errors
print("Mean squared errors for 50 regression models: mean = %.3f, std dev = %.3f" %(mean_mse, std_mse))
```

#B. Normalize the data (5 marks) 

Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

```python
x_norm = (x - x.mean()) / x.std()
y_norm = (y - y.mean()) / y.std()
```

```python
display(x_norm.head())
display(y_norm.head())
```

```python
model2 = reg_mod()
mean_mse, std_mse = evaluate_model(model2, x_norm, y_norm, 50)
# Report the mean and stddev of the mean squared errors
print("Mean squared errors for 50 regression models: mean = %.3f, std dev = %.3f" %(mean_mse, std_mse))
```
